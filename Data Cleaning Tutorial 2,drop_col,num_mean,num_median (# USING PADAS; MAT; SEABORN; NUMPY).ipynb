```python
null_var = df.isnull().sum()/df.shape[0]*100
null_var
```

```python
                                # coloums
```

```python
find_col = null_var[ null_var > 20 ].keys()
print(find_col)
```

```python
drop_col = df.drop( columns= find_col)
```

```python
df.shape                                      # original dataset
```

```python
drop_col.shape                           # after removing/ droping coloumns which is containt VERY LESS DATA
```

```python
plt.figure(figsize= (16,9))
sns.heatmap(drop_col.isnull())
plt.show()
```

```python
                              #  ONLY NUMERIC
```

```python
drop_col_num = drop_col.select_dtypes( include= ['int64','float64'])
drop_col_num.head()
```

```python
plt.figure(figsize= (16,9))
sns.heatmap(drop_col_num.isnull())
plt.show()
```

```python
drop_col_num.isnull().sum()
```

```python
drop_col_num[drop_col_num.isnull().any(axis = 1)]
drop_col_num.head()
```

```python
missing_num_val = [var for var in drop_col_num.columns 
                         if drop_col_num[var].isnull().sum() > 2]
missing_num_val
```

```python
plt.figure(figsize=(25,25))

sns.set()

for i, var in enumerate(missing_num_val):
    plt.subplot(2,2, i+1)
    
    sns.histplot( drop_col_num[var] , bins=20 ,edgecolor = "r" , kde=True, linewidth = 4 , linestyle = "--" ,alpha=0.2, color='#4682b4' ,line_kws={'lw': 10})
```

# MEAN

```python
num_mean = drop_col_num.fillna(drop_col_num.mean())
num_mean.isnull().sum().sum()
```

```python

plt.figure(figsize=(25,25))
#kde_kw=kde_kw

sns.set()

for i, var in enumerate(missing_num_val):
    plt.subplot(2,2, i+1)
    


    sns.histplot( data=drop_col_num[var] , bins=20 ,edgecolor = "y" , kde=True, linewidth = 5,  linestyle = "--" , alpha=0.1, color='k',line_kws={'lw': 6}) 
    sns.histplot( data=num_mean[var] , bins=10,edgecolor = "orange" , kde=True, linewidth = 4 , linestyle = "--" ,alpha=0.2, color='#4682b4' ,line_kws={'lw': 8})#line_kws=None,kde= True, kde_kws={ 'color':'green' , 'linewidth' :8 } )  
    
   # plt.legend()
```

# MEDIAN

```python
num_median = drop_col_num.fillna(drop_col_num.median())
num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(25,25))

sns.set()

print(missing_num_val)

for i, var in enumerate(missing_num_val):
    plt.subplot(2,2, i+1)
    

    sns.histplot( drop_col_num[var] , bins=20,edgecolor = "y" , kde=True, linewidth = 5,  linestyle = "--" , alpha=0.1, color='k',line_kws={'lw': 6})  
    sns.histplot( num_median[var] , bins=20 , edgecolor = "g",kde=True, linewidth = 4 , linestyle = "--" ,alpha=0.3, color='r' ,line_kws={'lw': 4} )
```

```python
plt.figure(figsize=(25,25))

sns.set()

for i, var in enumerate(missing_num_val):
    plt.subplot(2,2, i+1)
    

    sns.histplot( drop_col_num[var] , bins=20,edgecolor = "y" , kde=True, linewidth = 5,  linestyle = "--" , alpha=0.1, color='k',line_kws={'lw': 6} )  
    sns.histplot( num_mean[var] , bins=20,edgecolor = "orange" , kde=True, linewidth = 4 , linestyle = "--" ,alpha=0.2, color='#4682b4' ,line_kws={'lw': 8})
    sns.histplot( num_median[var] , bins=20 , edgecolor = "g",kde=True, linewidth = 4 , linestyle = "--" ,alpha=0.3, color='r' ,line_kws={'lw': 4} )
```

```python
for i, var in enumerate(missing_num_val):
    
    plt.figure(figsize=(10,10))
    
    plt.subplot(3,1,1)
    sns.boxplot( x= drop_col_num[var])
    
    plt.subplot(3,1,2)
    sns.boxplot(x= num_median[var])
   
    plt.subplot(3,1,3)
    sns.boxplot(x= num_mean[var])
```

```python
df_concat = pd.concat([drop_col_num[missing_num_val],num_mean[missing_num_val],num_median[missing_num_val]],axis=1)
```
