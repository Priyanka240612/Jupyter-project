```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# histogram always plot in one per labled value, if you have 2 lableled data colom must be draw seperete
# like shown in next
#  ml_stu_age
```

```python
ml_stu_age = np.random.randint(15,45,50)

print(ml_stu_age)
```

```python
ml_stu_age.max()
```

```python
plt.hist(ml_stu_age)
```

```python
plt.hist(ml_stu_age,bins=[5,10,15,20,25,30,35,40,45,50])
```

```python
plt.hist(ml_stu_age,bins=[5,10,15,20,25,30,35,40,45,50], rwidth= 0.8)
```

```python
plt.hist(ml_stu_age,bins=[5,10,15,20,25,30,35,40,45,50], rwidth= 0.8 , color = "y",label="ml_stu")


plt.title("Student Age", fontsize = 18)
plt.xlabel("Age category", fontsize = 15)
plt.ylabel("No. of stu ", fontsize = 15)


plt.legend(loc=2)
```

```python
plt.hist(ml_stu_age,bins=[5,10,15,20,25,30,35,40,45,50], rwidth= 0.8 , color = "m",label="ml_stu", orientation='horizontal')


plt.title("Student Age", fontsize = 18)
plt.xlabel("Age category", fontsize = 15)
plt.ylabel("No. of stu ", fontsize = 15)


plt.legend(loc=2)
```

```python
plt.hist(ml_stu_age,bins=[5,10,15,20,25,30,35,40,45,50], rwidth= 0.8 , color = "b",label="ml_stu",orientation='horizontal',
        histtype = "stepfilled")


plt.title("Student Age", fontsize = 18)
plt.xlabel("Age category", fontsize = 15)
plt.ylabel("No. of stu ", fontsize = 15)


plt.legend(loc=2)
```

```python
plt.hist(ml_stu_age,bins=[5,10,15,20,25,30,35,40,45,50], rwidth= 0.8 , color = "b",label="ml_stu",orientation='horizontal',
        histtype = "step")


plt.title("Student Age", fontsize = 18)
plt.xlabel("Age category", fontsize = 15)
plt.ylabel("No. of stu ", fontsize = 15)

plt.legend(loc=2)
```

```python
plt.hist(ml_stu_age,bins=[5,10,15,20,25,30,35,40,45,50], rwidth= 0.8 , color = "k",label="ml_stu",histtype = "bar")


plt.title("Student Age", fontsize = 18)
plt.xlabel("Age category", fontsize = 15)
plt.ylabel("No. of stu ", fontsize = 15)

plt.legend(loc=2)
```

```python
py_stu_age = np.random.randint(12,60,50)
print(py_stu_age)
```

```python
plt.hist(ml_stu_age, rwidth= 0.8 , color = "r",label="ml_stu")
plt.hist(py_stu_age, rwidth= 0.8 , color = "y",label="py_stu")


plt.title("Student Age", fontsize = 18)
plt.xlabel("Age category", fontsize = 15)
plt.ylabel("No. of stu ", fontsize = 15)

plt.legend(loc=2)
```

```python
plt.hist(ml_stu_age, rwidth= 0.8 , color = "r",label="ml_stu",)
plt.hist(py_stu_age, rwidth= 0.8 , color = "y",label="py_stu",orientation='horizontal')

plt.title("Student Age", fontsize = 18)
plt.xlabel("Age category", fontsize = 15)
plt.ylabel("No. of stu ", fontsize = 15)

plt.legend(loc=0)
```

```python
plt.hist([ml_stu_age,py_stu_age], rwidth= 0.8 , color = ["g","y"],label=["ml_stu","py_stu"])


plt.title("Student Age", fontsize = 18)
plt.xlabel("Age category", fontsize = 15)
plt.ylabel("No. of stu ", fontsize = 15)

plt.legend(loc=2)
```

```python

plt.figure( figsize = (15,5))

plt.hist([ml_stu_age,py_stu_age], rwidth= 0.8 , color = ["g","y"],label=["ml_stu","py_stu"])


plt.title("Student Age", fontsize = 18)
plt.xlabel("Age category", fontsize = 15)
plt.ylabel("No. of stu ", fontsize = 15)

plt.legend(loc=2)
```

```python
from matplotlib import style
style.use("ggplot")

plt.figure( figsize = (15,5))

plt.hist([ml_stu_age,py_stu_age], rwidth= 0.8 , color = ["r","y"],label=["ml_stu","py_stu"])


plt.title("Student Age", fontsize = 18)
plt.xlabel("Age category", fontsize = 15)
plt.ylabel("No. of stu ", fontsize = 15)

plt.legend(loc=2)
```
