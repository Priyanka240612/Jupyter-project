```python
def catog_var(obj_var):
    return pd.concat( [df[obj_var].value_counts() / df.shape[0]*100,
           drop_row[obj_var].value_counts() / drop_row.shape[0]*100],
           axis =1 , keys = [obj_var+'_org',obj_var+'_new'])
```
