```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_diabetes
diabetes_df = load_diabetes()
diabetes_df
```

```python
diabetes_df1 = pd.DataFrame(np.c_[diabetes_df['data'],diabetes_df['target']],columns=np.append(diabetes_df['feature_names'],['target']))
diabetes_df1
```

```python
sns.pairplot(diabetes_df1 )
plt.show()
```

```python
sns.pairplot(diabetes_df1 ,x_vars='age', hue='target')
plt.show()

# hue=None,
#  hue_order=None,
#    palette=None,
 #   vars=None,
  #  x_vars=None,
   # y_vars=None,
```

```python
sns.pairplot(diabetes_df1 ,y_vars='age')
plt.show()
```

```python
sns.pairplot(diabetes_df1 ,y_vars='sex' , hue='age')
plt.show()
```

```python
sns.pairplot(diabetes_df1 ,y_vars='sex' , hue='age', hue_order= [1.0, 0.0])
plt.show()
```
