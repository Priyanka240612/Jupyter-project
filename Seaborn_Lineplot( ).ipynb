```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
days =[1,2,3,4,5,6,7,8,9]
temperature =[31,31.5,34,33,35,37,36,34.5,35]

wheather_df = pd.DataFrame({"days":days, "temperature":temperature })
wheather_df
```

```python
sns.lineplot(x = days,y = temperature, data = wheather_df )

plt.title("wheather"  , fontsize = 18)
plt.xlabel("days",fontsize = 15)
plt.ylabel("temperature", fontsize = 15)

plt.show()
```
