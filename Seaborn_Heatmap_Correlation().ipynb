```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#import sklearn 
```

```python
glob_warmg = pd.read_csv('C:\Workspace\Pri\data science and ML practice\jupyter\DataBases\Who_is_responsible_for_global_warming.csv')
glob_warmg.head(10)
```

```python
glob_warmg_df = glob_warmg.corr( numeric_only = True )   
glob_warmg_df
```

```python
plt.figure(figsize=(16,9))

hm = sns.heatmap( data = glob_warmg_df , annot= True, linewidths = 4)

hm.tick_params(size = 10, color = 'k', labelsize= 10 ,labelcolor = 'b')

plt.title(" GLOBAL WARMING")

plt.show()
```

```python
from sklearn.datasets import load_diabetes
diabetes_df = load_diabetes()
diabetes_df
```

```python
diabetes_df1 = pd.DataFrame(np.c_[diabetes_df['data'],diabetes_df['target']],columns=np.append(diabetes_df['feature_names'],['target']))
diabetes_df1
```

```python
plt.figure(figsize=(16,9))

hm = sns.heatmap( data = diabetes_df1.corr() , annot= True, linewidths = 4)

hm.tick_params(size = 10, color = 'k', labelsize= 10 ,labelcolor = 'b')

plt.title(" DIABETES")

plt.show()
```
