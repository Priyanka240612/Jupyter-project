# Its convenient method for combining the COLUMNS of two potensially different indexes

# Syntax : DataFrame.join()     * df.join()

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'ID':[101,102,103,104,105],'Class':[1,2,3,4,5],'Name':['A','B','C','D','E'] })
df1

df2= pd.DataFrame({'Marks':[106,107,108,109,1010],'Standard':[6,7,8,9,10], 'Grade':['F','G','H','I','J']})
df2

df3 = pd.DataFrame({'Pages':[405,406,407,408,409],'Pen':[5,6,7,8,9], 'Point':['D','E','G','H','I']} ,
                  index = ['z','y','x','w','v'])
df3

df4 = pd.DataFrame({'ID':[101,102,103,104,105],'Class':[1,2,3,4,5],'Name':['A','B','C','D','E']} ,
                  index = ['z','y','x','w','v'])
df4

df5 = pd.DataFrame({'Items':[101,102,103],'Cake':[1,2,3],'Ice-cream':['A','B','C']} ,
                  index = ['z','y','x'])
df5

df6 = pd.DataFrame({'Items':[101,102,103],'Cake':[1,2,3],'Ice-cream':['A','B','C']} ,
                  index = ['z','y','x'])
df6

display(df1, df2 , df3 , df4 , df5 ,df6)
```

```python
J1 = df1.join(df2)
J1
```

```python
J2 = df2.join(df1)
J2
```
