```python
plt.figure(figsize=(15,10))

plt.subplot(3,2,1)

days =[1,2,3,4,5,6,7,8,9]
Mumbai_temp =[31,31.5,34,33,35,37,36,34.5,35]
Goa_temp =[32,33.5,35,34,36,38,36,35.5,36]
plt.plot(days,Mumbai_temp, 'yo--',markersize = 10 , linewidth = 3 ,label = "Temp line") 
plt.plot(days,Goa_temp, 'm^-',markersize = 10 , linewidth = 3, label = "Temp line")
plt.legend( loc = 2)
plt.xlabel("days",fontsize = 15)
plt.ylabel("temperature", fontsize = 15)
import matplotlib.style as style
style.use("ggplot")
plt.grid(color = 'w', linewidth = 2 , linestyle = '-')



plt.subplot(3,2,2)

ml_stu_age = np.random.randint(15,45,50)
py_stu_age = np.random.randint(12,60,50)
style.use("ggplot")
#plt.figure( figsize = (15,5))
plt.hist([ml_stu_age,py_stu_age], rwidth= 0.8 , color = ["r","y"],label=["ml_stu","py_stu"])
plt.title("Student Age", fontsize = 18)
plt.xlabel("Age category", fontsize = 15)
plt.ylabel("No. of stu ", fontsize = 15)
plt.legend(loc=2)




plt.subplot(3,2,3)

df = pd.read_csv('C:\Workspace\Pri\data science and ML practice\jupyter\DataBases\Spotify_Youtube.csv', nrows =50)
#df
x = df["Views"]
y = df["Likes"]
#plt.scatter( x ,y )
plt.scatter( x ,y , s= 100 ,marker = "*",alpha = 0.7, linewidth = 14 , edgecolor = "y" )
plt.title("Spotify_Youtube" , fontsize = 18 )
plt.xlabel("Views", fontsize = 15 )
plt.ylabel("Likes", fontsize = 15 )




plt.subplot(324)

Batches = ["PYTHON","R","ML","AI","DS"]
stu1_marks = [30,20,35,40,45]
stu2_marks = [34,25,34,44,48]
stu3_marks = [33,28,39,49,41]
Batches_index = np.arange(len(Batches))
width = 0.2
plt.barh( Batches_index  ,stu1_marks,     width,    color = "y",label ="stu1_marks")
plt.barh(Batches_index + width  ,stu2_marks,     width,    color = "m",label ="stu2_marks")
plt.barh(Batches_index + width + width ,stu3_marks,    width,    color = "g",label ="stu3_marks")
#  after passing width inside command its take value
plt.xticks ( Batches_index + width , Batches , rotation = 25 )
plt.legend()
plt.title("Student Result", fontsize = 18)
plt.xlabel("Batches category", fontsize = 15)
plt.ylabel("Student marks", fontsize = 15)


plt.subplot(325)


Batches = ["PYTHON","R","ML","AI","DS"]
stu1_marks = [30,20,35,40,45]
explode=[0,0,0.2,0,0]
wedgeprops = { "linewidth": 4 , "width": 1 , "edgecolor": "k"}
plt.pie(stu1_marks , labels = Batches ,   wedgeprops = wedgeprops , explode=explode , radius= 1.4 )
#plt.legend()


plt.subplot(3,2,6 , projection = 'hammer' , facecolor = 'm') 
plt.show()
# projetion : {None, 'aitoff', 'hammer', 'lambert', 'mollweide', 'polar', 'rectilinear', str}, optional


```
