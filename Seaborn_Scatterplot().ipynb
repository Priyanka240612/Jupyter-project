```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
tita_df = sns.load_dataset("titanic")
tita_df
```

```python
sns.scatterplot(data= tita_df, x = "age", y = "fare" ,hue = "sex")
plt.show()
```

```python
sns.scatterplot(data= tita_df, x = "age", y = "fare" ,hue = "sex" , style= 'who')
plt.show()
```

```python
sns.scatterplot(data= tita_df, x = "age", y = "fare" ,hue = "sex" , style= 'who' , size ="who")
plt.show()
```

```python
sns.scatterplot(data= tita_df, x = "age", y = "fare" ,hue = "sex" , style= 'who' , size ="who" , sizes = (100,400))
plt.show()
```

```python
sns.scatterplot(data= tita_df, x = "who", y = "fare" ,hue = "alive" , style= 'alive' , size ="who" , sizes = (100,400))
plt.show()
```

```python
sns.scatterplot(data= tita_df, x = "who", y = "fare" ,hue = "alive" , style= 'alive' , size ="who" , sizes = (100,400),
                 palette= "Accent" )
plt.show()
```

```python
sns.scatterplot(data= tita_df, x = "who", y = "fare" ,hue = "alive" , style= 'alive' , size ="who" , sizes = (100,400),
                 palette= "Accent" , alpha = .7)
plt.show()
```

```python
plt.figure( figsize = (10,9))


sns.scatterplot(data= tita_df, x = "age", y = "fare" )
sns.lineplot(data= tita_df, x = "age", y = "fare" )
sns.barplot(data= tita_df, x = "age", y = "fare" )


plt.show()
```
