```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips_df = sns.load_dataset("tips")
tips_df
```

```python
plt.figure( figsize = (15,9))

sns.barplot( data=tips_df, x ="tip", y ="total_bill")

plt.show()
```

```python
plt.figure( figsize = (15,9))

sns.barplot( data=tips_df,x ="day", y ="total_bill")

plt.show()
```

```python
#plt.figure( figsize = (15,9))

sns.barplot( data=tips_df,x ="day", y ="total_bill" ,hue= "sex")

plt.show()
```

```python
hue_order=['Female', 'Male']

sns.barplot( data=tips_df,x ="day", y ="total_bill" ,hue= "sex" , hue_order=hue_order)

plt.show()
```

```python
order = ['Sun', 'Fri', 'Thur', 'Sat']


sns.barplot( data=tips_df,x ="day", y ="total_bill" , order =order )

plt.show()
```

```python
hue_order=['Female', 'Male']

sns.barplot( data=tips_df,x ="day", y ="total_bill" ,hue= "sex" , hue_order=hue_order , estimator='max') 
                                                                                    # estimator='mean' default value

plt.show()
```

```python
hue_order=['Female', 'Male']

sns.barplot( data=tips_df,x ="day", y ="total_bill" ,hue= "sex" , hue_order=hue_order , n_boot=2) 
                                                                                    # n_boot=1000 default
plt.show()
```

```python
ax = sns.barplot( data=tips_df,x ="day", y ="total_bill" ,hue= "sex")

ax.set( title  = " Tips Github", 
        xlabel = " Days",
        ylabel = " Total_bill "
        )

plt.show()
```

```python
sns.barplot( data=tips_df,x ="day", y ="total_bill" ,hue= "sex")

plt.title("Student Result", fontsize = 18)
plt.xlabel("Batches category", fontsize = 15)
plt.ylabel("Student marks", fontsize = 15)

plt.show()
```
