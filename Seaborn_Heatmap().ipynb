```python
glob_warmg = pd.read_csv('C:\Workspace\Pri\data science and ML practice\jupyter\DataBases\Who_is_responsible_for_global_warming.csv')
glob_warmg.head(10)
```

```python
glob_warmg1 = glob_warmg.drop(columns = ['Country Code', 'Indicator Name', 'Indicator Code'] , axis= 1).set_index('Country Name')
glob_warmg1.head(10)
```

```python
plt.figure(figsize= (16,9))

sns.heatmap(glob_warmg1)

plt.show()
```

```python
plt.figure(figsize= (16,9))
sns.heatmap(glob_warmg1,  vmin=0, vmax=10 )
```

```python
plt.figure(figsize= (16,9))

sns.heatmap(glob_warmg1 , cmap= "hot")

plt.show()
```

```python
plt.figure(figsize= (16,9))

sns.heatmap(glob_warmg1 , cmap= "hot" , annot= True )

plt.show()
```

```python
plt.figure(figsize= (16,9))

sns.heatmap(glob_warmg1 , cmap= "hot" , annot= True ,fmt='0.5')

plt.show()
```

```python
arr_stri = np.array([['aa','bb','cc'] , ['dd','ee','ff'] , ['gg','hh','ii'] , ['jj','kk','ll']])
arr_stri
```

```python
plt.figure(figsize= (16,9))

sns.heatmap( data = arr ,annot = arr_stri , cmap= "Accent" ,fmt='s')

plt.show()
```

```python
plt.figure(figsize= (16,9))

sns.heatmap( data = arr ,annot = arr_stri , cmap= "Accent" ,fmt='s' , linewidths = 8 , linecolor='k')
                                                              # default ** linewidths=0 ,  linecolor='white'
plt.show()
```

```python
sns.heatmap( data = arr ,annot = arr_stri , cmap= "Accent" ,fmt='s' , linewidths = 8 , linecolor='k', xticklabels= False,
             yticklabels= False,  cbar= False)           # default ** xticklabels='auto', yticklabels='auto',  cbar=True 
                                                                
plt.show()
```

```python
cbar_kws= {"orientation":"horizontal"}

sns.heatmap( data = arr ,annot = arr_stri , cmap= "Accent" ,fmt='s' , linewidths = 8 , linecolor='k' , cbar_kws=cbar_kws)

plt.show()
                                                              
```
