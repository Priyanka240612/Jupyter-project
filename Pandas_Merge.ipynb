```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'ID':[101,102,103,104,105],'Class':[1,2,3,4,5]})
df1
```

```python
df2= pd.DataFrame({'ID':[101,102,103,104,105], 'Name':['A','B','C','D','E']})
df2
```

```python
df3 = pd.DataFrame({'ID':[101,102,103,106,107], 'Section':['a','b','c','d','e']})
df3
```

```python
M1 = pd.merge(df1,df2)     #      or     M1 = pd.merge(df1,df2, on= 'ID' )
M1
```

```python
M2 = pd.merge(df2,df1 )     #    or    M2 = pd.merge(df2,df1 , on = 'ID' )
M2
```

```python
M3 = pd.merge(df1,df3)
M3
```

```python
M3a = pd.merge(df1,df3 , on='ID' , how = 'inner')
M3a
```

```python
M3a = pd.merge(df1,df3 , on='ID' , how = 'outer')
M3a
```

```python
M3aa = pd.merge(df1,df3 , on='ID' , how = 'outer', indicator= True)
M3aa
```

```python
M3b = pd.merge(df1,df3 , on='ID' , how = 'left')
M3b
```

```python
M3c = pd.merge(df1,df3 , on='ID' , how = 'right')
M3c
```
