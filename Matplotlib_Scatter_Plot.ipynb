```python
df = pd.read_csv('C:\Workspace\Pri\data science and ML practice\jupyter\DataBases\Spotify_Youtube.csv', nrows = 100)
```

```python

#  plt.scatter(Views,Likes )

#error  =  name 'Views' is not defined


x = df["Views"]
y = df["Likes"]

plt.scatter( x ,y )
```

```python
plt.figure(figsize= (16,9))
plt.scatter( x ,y , c = "m", s= 100 ,marker = "*",alpha = 0.7, linewidth = 14 , edgecolor = "y" )

plt.title("Spotify_Youtube" , fontsize = 18 )
plt.xlabel("Views", fontsize = 15 , rotation = 20)
plt.ylabel("Likes", fontsize = 15 , rotation = 20)
plt.show()
```

```python
z = df["Comments"]

plt.figure(figsize= (16,9))

plt.scatter( x ,y , c = "m", s= 100 ,marker = "*",alpha = 0.7, linewidth = 14 , edgecolor = "y" )

plt.scatter( x ,z , c = "k", s= 100 ,marker = "<",alpha = 0.7, linewidth = 14 , edgecolor = "g" )

plt.title("Spotify_Youtube" , fontsize = 18 )
plt.xlabel("Views", fontsize = 15 , rotation = 25)
plt.ylabel("Likes and Comments", fontsize = 15 )
plt.show()
```
