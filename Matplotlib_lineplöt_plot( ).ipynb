# Line Plot

```python
import matplotlib.pyplot as plt
```

```python
days =[1,2,3,4,5,6,7,8,9]
temperature =[31,31.5,34,33,35,37,36,34.5,35]

plt.plot(days,temperature)
plt.show()
```

```python
days =[1,2,3,4,5,6,7,8,9]
temperature =[31,31.5,34,33,35,37,36,34.5,35]

plt.plot(days,temperature)

plt.title("Mumbai Temperature")
plt.xlabel("days")
plt.ylabel("temperature")

plt.show()
```

```python
days =[1,2,3,4,5,6,7,8,9]
temperature =[31,31.5,34,33,35,37,36,34.5,35]

plt.plot(days,temperature)

plt.axis([0,10,0,40])

plt.title("Mumbai Temperature")
plt.xlabel("days")
plt.ylabel("temperature")

plt.show()
```

```python
days =[1,2,3,4,5,6,7,8,9]
temperature =[31,31.5,34,33,35,37,36,34.5,35]

plt.plot(days,temperature,   color = 'r' )

plt.axis([0,10,0,40])

plt.title("Mumbai Temperature")
plt.xlabel("days")
plt.ylabel("temperature")

plt.show()
```

```python
days =[1,2,3,4,5,6,7,8,9]
temperature =[31,31.5,34,33,35,37,36,34.5,35]

plt.plot(days,temperature,   color = 'y' , marker = 'o' ,markersize = 10,  linestyle = '--' , linewidth = 3)

plt.axis([0,10,0,40])

plt.title("Mumbai Temperature")
plt.xlabel("days")
plt.ylabel("temperature")

plt.show()
```

```python
days =[1,2,3,4,5,6,7,8,9]
temperature =[31,31.5,34,33,35,37,36,34.5,35]

plt.plot(days,temperature, 'yo--',markersize = 10 , linewidth = 3)   #  cut the command please see in the middle *color,shape,linestyle

plt.axis([0,10,0,40])

plt.title("Mumbai Temperature")
plt.xlabel("days")
plt.ylabel("temperature")

plt.show()
```

```python
days =[1,2,3,4,5,6,7,8,9]
Mumbai_temp =[31,31.5,34,33,35,37,36,34.5,35]
Goa_temp =[32,33.5,35,34,36,38,36,35.5,36]

plt.plot(days,Mumbai_temp, 'yo--',markersize = 10 , linewidth = 3 ,label = "Temp line") 
plt.plot(days,Goa_temp, 'm^-',markersize = 10 , linewidth = 3, label = "Temp line")
plt.legend( loc = 2)

# plt.axis([0,10,0,40])

plt.title("Temperature Graph"  , fontsize = 18)
plt.xlabel("days",fontsize = 15)
plt.ylabel("temperature", fontsize = 15)





import matplotlib.style as style
style.use("ggplot")
plt.grid(color = 'w', linewidth = 2 , linestyle = '-')

plt.show()
```
